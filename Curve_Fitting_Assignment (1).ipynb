```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as so
import math as m
```

```python
x1,y1,z1=np.loadtxt("data1.txt",unpack=True)
x2,y2,z2=np.loadtxt("data2.txt",unpack=True)
x3,y3,z3=np.loadtxt("data3.txt",unpack=True)
```

```python
def gauss(x, amp, cen, wid):
    return [amp*m.exp(-(xi-cen)**2 / wid) for xi in x]

def s(x,a,k,ph):
    return [a*m.sin(k*xi+ph) for xi in x]

def four(x,a,b,c,d,e):
    return a*x**4+b*x**3+c*x**2+d*x+e 

def sec(x,a,b,c):
    return a*x**2+b*x+c

def cube(x,a,b,c,d):
    return a*x**3+b*x**2+c*x+d

def e(x,a,b,c):
    return [a*np.exp(b*xi)+c for xi in x]
```

```python
popt1,pcov1=so.curve_fit(gauss,x1,y1,sigma=z1)

yf1=gauss(x1,popt1[0],popt1[1],popt1[2])

ym=np.mean(y1)
ssres=np.sum((y1-yf1)**2)
sstot=np.sum((y1-ym)**2)
gf1=1-(ssres/sstot)
print("Goodness of fit for Gaussian fit on the first data set is: "+str(gf1))

popt2,pcov2=so.curve_fit(s,x1,y1,sigma=z1)

yf2=s(x1,popt2[0],popt2[1],popt2[2])

ym=np.mean(y1)
ssres=np.sum((y1-yf2)**2)
sstot=np.sum((y1-ym)**2)
gf2=1-(ssres/sstot)
print("Goodness of fit for the Sine fit on the first data set is: "+str(gf2))

popt3,pcov3=so.curve_fit(s,x1,y1,sigma=z1)

yf3=e(x1,*popt3)

ym=np.mean(y1)
ssres=np.sum((y1-yf3)**2)
sstot=np.sum((y1-ym)**2)
gf3=1-(ssres/sstot)
print("Goodness of fit for the exponential fit on the first data set is: "+str(gf3))
```

```python
plt.figure()
plt.errorbar(x1,y1,xerr=None,yerr=z1,ecolor='gray',capsize=1,barsabove=True,label="Original",ls='--')
plt.plot(x1,yf1,color="red",marker='.',label="Fitted")
plt.title("Fitting Gaussian curve on first data set")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
plt.plot(x1,yf2,marker='.',label='fitted',color='red')
plt.errorbar(x1,y1,xerr=None,yerr=z1,ecolor='gray',capsize=1,barsabove=True,label="Original",ls='--')
plt.title("Fitting Sine curve on first data set")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
plt.figure()
plt.errorbar(x1,y1,xerr=None,yerr=z1,ecolor='gray',capsize=1,barsabove=True,label="Original",ls='--')
plt.plot(x1,yf3,color="red",marker='.',label="Fitted")
plt.title("Fitting exponential curve on first data set")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
popt1,pcov1=so.curve_fit(s,x2,y2,sigma=z2)

yf1=s(x2,popt1[0],popt1[1],popt1[2])

ym=np.mean(y2)
ssres=np.sum((y2-yf1)**2)
sstot=np.sum((y2-ym)**2)
gf1=1-(ssres/sstot)
print("Goodness of fit for sine curve on second data set is: "+str(gf1))

c1=np.polyfit(x2,y2,4)

yf2=four(x2,c1[0],c1[1],c1[2],c1[3],c1[4])

ym=np.mean(y2)
ssres=np.sum((y2-yf2)**2)
sstot=np.sum((y2-ym)**2)
gf2=1-(ssres/sstot)
print("Goodness of fit for 4th degree polynomial on the second data set is: "+str(gf2))

c2=np.polyfit(x2,y2,3)

yf3=cube(x2,*c2)

ym=np.mean(y2)
ssres=np.sum((y2-yf3)**2)
sstot=np.sum((y2-ym)**2)
gf3=1-(ssres/sstot)
print("Goodness of fit for cubic polynomial on the second data set is: "+str(gf3))
```

```python
plt.plot(x2,yf1,marker='.',label='fitted',color='red')
plt.errorbar(x2,y2,xerr=None,yerr=z2,ecolor='gray',capsize=1,barsabove=True,label="Original",ls='--')
plt.title("Fitting Sine curve on second data set")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
plt.plot(x2,yf2,marker='.',label='fitted',color='red')
plt.errorbar(x2,y2,xerr=None,yerr=z2,ecolor='gray',capsize=1,barsabove=True,label="Original",ls='--')
plt.title("Fitting 4th degree polynomial on second data set")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
plt.plot(x2,yf3,marker='.',label='fitted',color='red')
plt.errorbar(x2,y2,xerr=None,yerr=z2,ecolor='gray',capsize=1,barsabove=True,label="Original",ls='--')
plt.title("Fitting cubic polynomial on second data set")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
popt1,pcov1=so.curve_fit(sec,x3,y3,sigma=z3)

yf1=sec(x3,popt1[0],popt1[1],popt1[2])

ym=np.mean(y3)
ssres=np.sum((y3-yf1)**2)
sstot=np.sum((y3-ym)**2)
gf1=1-(ssres/sstot)
print("Goodness of fit for a quadratic polynomial on the third data set is: "+str(gf1))

popt1,pcov1=so.curve_fit(cube,x3,y3,sigma=z3)

yf2=cube(x3,popt1[0],popt1[1],popt1[2],popt1[3])

ym=np.mean(y3)
ssres=np.sum((y3-yf2)**2)
sstot=np.sum((y3-ym)**2)
gf2=1-(ssres/sstot)
print("Goodness of fit for a cubic polynomial on the third set of data is: "+str(gf2))

popt1,pcov1=so.curve_fit(four,x3,y3,sigma=z3)

yf3=four(x3,*popt1)

ym=np.mean(y3)
ssres=np.sum((y3-yf3)**2)
sstot=np.sum((y3-ym)**2)
gf3=1-(ssres/sstot)
print("Goodness of fit for a 4th degree polynomial on the third set of data is: "+str(gf3))
```

```python
plt.plot(x3,yf1,marker='.',label='fitted',color='red')
plt.errorbar(x3,y3,xerr=None,yerr=z3,ecolor='gray',capsize=1,barsabove=True,label="Original",ls='--')
plt.title("Fitting quadratic polynomial on third data set")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
plt.plot(x3,yf2,marker='.',label='fitted',color='red')
plt.errorbar(x3,y3,xerr=None,yerr=z3,ecolor='gray',capsize=1,barsabove=True,label="Original",ls='--')
plt.title("Fitting cubic polynomial on third data set")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
plt.plot(x3,yf2,marker='.',label='fitted',color='red')
plt.errorbar(x3,y3,xerr=None,yerr=z3,ecolor='gray',capsize=1,barsabove=True,label="Original",ls='--')
plt.title("Fitting 4th degree polynomial on third data set")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
print("To conclude: \n 1. For first data set, Gaussian function is the best fit \n 2. For second data set, 4th order polynomial is the best fit \n 3. For third data set, quadratic polynomial is the best fit.")
```
