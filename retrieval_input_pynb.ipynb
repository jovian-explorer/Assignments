```python
#importing libraries required
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/jovian-explorer/petitRADTRANS/master/test.csv'

pd.set_option('display.max_columns', None) 
pd.set_option('display.max_rows', None)  
pd.set_option('display.max_colwidth', None)

dataset = pd.read_csv(url)
# print(dataset)
```

```python
# accessing the different columns of the csv dataset I made above
dataset.columns = ["wavelength","bin_width","tr_depth","tr_depth_errneg","tr_depth_errpos","ecl_depth", "ecl_depth_errneg", "ecl_depth_errpos"]
#storing values for each column into a separate arrays
wavelength = np.array(dataset.wavelength)            # Values of wavelength
tr_depth = np.array(dataset.tr_depth)                # Values of transit depth
tr_depth_errneg = np.array(dataset.tr_depth_errneg)  # Values of tr_depth_errneg
tr_depth_errpos = np.array(dataset.tr_depth_errpos)  # Values of tr_depth_errpos

plt.scatter(wavelength, tr_depth_errneg, s = 10,label = "Error neg")
plt.scatter(wavelength, tr_depth_errpos, s = 10,color = 'r',label = "Error pos")

plt.title("Graph between Wavelength and Transit Depth Errors")
plt.legend()
plt.show()

plt.scatter(wavelength, tr_depth, s = 10,label = "Transit depth")
plt.plot(wavelength, tr_depth,label = "Transit depth")

plt.title("Graph between Wavelength and Transit Depth")
plt.legend()
plt.show()
```
