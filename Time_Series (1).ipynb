```python
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.tsa.stattools as ss
```

```python
'''
s: the 1d array containing the data that needs to be checked for stationarity
sig: the significance level
label: a string that describes what data is being provided
'''
t=[]
def diff(s,sig,label):
    diffd=[]
    for i in range(0,len(s)-1):
        diffd.append(s[i+1]-s[i])
    plt.figure(1)
    plt.plot(diffd)
    plt.title(str(label)+" after differencing")
    res=ss.adfuller(diffd)
    if res[1]>sig:
        print("Differencing didn't work on "+str(label))
        print("p-Value is: "+str(res[1]))
    elif res[1]<=sig:
        print(str(label)+" has been made stationary through differencing")
        print("p=Value is: "+str(res[1]))
    t.append(diffd)
    return

'''
s: the 1d array containing the data that needs to be checked for stationarity
sig: the significance level
label: a string that describes what data is being provided
'''
def isStat(s,sig,label):
    plt.figure(0)
    plt.plot(s)
    plt.title(str(label))
    res=ss.adfuller(s)
    if res[1]>sig:
        print("Null hypothesis could not be rejected. "+str(label)+" is non-stationary.")
        print("p-Value is: "+str(res[1]))
        diff(s,sig,label)
    elif res[1]<=sig:
        print("Null hypothesis rejected. "+str(label)+" is a stationary time series.")
        print("p=Value is: "+str(res[1]))
    return
```

```python
s1=np.loadtxt("AGN_LC.txt",usecols=[1],unpack=True)
s2,s3,s4,s5,s6=np.loadtxt("covid19_TS.txt",usecols=(1,2,3,4,5),unpack=True,skiprows=1)
```

```python
significance=0.05
isStat(s1,significance,"AGN Light Curve data")
```

```python
isStat(s2,significance,"Covid daily cases")
```

```python
isStat(s3,significance,"Covid total cases")
```

```python
isStat(s4,significance,"Covid daily recovered")
```

```python
isStat(s4,significance,"Covid total recovered")
```
